## Objective:
Creating a csv file that represents the relationship between the segments and the rivers.

```python
import geopandas as gpd
import pandas as pd
import fiona
from ipyfilechooser import FileChooser
from shapely.geometry import Polygon, LineString, Point
from geopandas import GeoSeries
```

```python
fc = FileChooser()
display(fc)
```

```python
file_path = fc.selected
fl_df = gpd.read_file(file_path)
fl_df.info()
```

## Creating segment ID
As the ids are not consistent, we use the row number and the DATASETNAME to create an ID for each row.

```python
fl_df['row_index'] = fl_df.index
fl_df['row_index'] = fl_df['row_index'].astype('str')
fl_df['segment_id'] = fl_df['DATASETNAM'] + "-" + fl_df.row_index
fl_df.head()
```

```python
rv_df = fl_df.dropna(subset=['NAME_1'])
rv_df = rv_df[['segment_id', 'NAME_1']].reset_index(drop=True)
rv_df.head()
```

```python
rv_df = pd.DataFrame(rv_df)
rv_df.columns = ['segment_id' , "river_name"]
```

```python
rv_df.river_name = rv_df.river_name.str.replace(' ','-')
```

```python
rv_df.to_csv("rivers_segments.csv")
```
